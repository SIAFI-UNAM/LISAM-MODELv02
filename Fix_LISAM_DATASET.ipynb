```python
import os
dataset_path = 'Object/'

dataset_txt_files = [file for file in os.listdir(dataset_path) if file.endswith(".txt")]
len(dataset_txt_files)
```

```python
def get_classes_dict(classes_file: str) -> dict:
  classes = {}

  with open(classes_file) as classes_file:
    data = classes_file.read()
    classes = {value: str(index) for index, value in enumerate(
      data.split()
    )}
  
  return classes
```

```python
def fix_class(file_path: str, classes: dict):
  letter = file_path.split("/")[-1].split('_')[0]
  content = []
  with open(file_path, 'r+') as txt_file:
    content = txt_file.read().split()
    content[0] = classes[letter]

  with open(file_path, 'w') as txt_file:
    txt_file.write(' '.join(content))
```

```python
classes = get_classes_dict('classes.txt')
classes
```

```python
errors = []
len_dataset = len(dataset_txt_files)
for index, txt_file in enumerate(dataset_txt_files):
  try:
    file_path = f'Object/{txt_file}'
    fix_class(file_path, classes)
  except:
    errors.append(txt_file)

  percentage = int((index / len_dataset) * 100)
  print(f'\rFixing Dataset {percentage}%', end='')
```
