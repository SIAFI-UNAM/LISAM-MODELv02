```python
test_path = f'{data_path}/test.txt'
train_path = f'{data_path}/train.txt'
# generate obj.data
objdata = {
    'classes': len(classes),
    'names': classes_file_path,
    'backup': training_path,
    'train': train_path,
    'test_path': test_path
}

objdata
```

```python
objdata_path = f'{data_path}/obj.data'

# generate obj.data
objdata_str = ''
for k, v in objdata.items():
  objdata_str += f'{k} = {v}\n'

text_file = open(objdata_path, "w")
n = text_file.write(objdata_str)
text_file.close()
```
