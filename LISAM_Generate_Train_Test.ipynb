```python
import glob, os

def generate_test_train(
    data_path: str, 
    output_path: str,
    percentage_test: int):
  # Create and/or truncate train.txt and test.txt
  file_train = open(f'{output_path}/train.txt', 'w')
  file_test = open(f'{output_path}/test.txt', 'w')

  # Populate train.txt and test.txt
  counter = 1
  index_test = round(100 / percentage_test)
  for pathAndFilename in glob.iglob(os.path.join(data_path, "*.jpg")):
      title, ext = os.path.splitext(os.path.basename(pathAndFilename))

      if counter == index_test:
          counter = 1
          file_test.write(data_path + "/" + title + '.jpg' + "\n")
      else:
          file_train.write(data_path + "/" + title + '.jpg' + "\n")
          counter = counter + 1
```

```python
dataset_path = f'/content/data/content/content/Object'
generate_test_train(dataset_path, data_path, 2)
```
